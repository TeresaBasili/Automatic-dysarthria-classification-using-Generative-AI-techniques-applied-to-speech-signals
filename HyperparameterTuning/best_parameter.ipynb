```python
import os
import json
import pandas as pd
```

```python
path='' #

all_data = []

# itera sui tre modelli
for model_name in os.listdir(path):
    model_path = os.path.join(path, model_name)
    # itera sul numero di combinazioni
    for iteraction in os.listdir(model_path):
        iter_path = os.path.join(model_path, iteraction)
        # itera sui file
        for file_name in os.listdir(iter_path):
            # prende solo il file.json, lo legge, copia i dati e aggiunge modello e n. combinazione
            if file_name.endswith('.json'):
                json_path = os.path.join(iter_path, file_name)
                with open(json_path, 'r') as f:
                    data = json.load(f)
                data['model'] = model_name
                data['iteraction'] = iteraction
                all_data.append(data)

df = pd.DataFrame(all_data)
```

```python
df.head(75)
```

```python
path_best_folder=''
models=[ 'resnet152', 'resnet50' ] #'mobilenet',
for model in models:
    if not os.path.exists(os.path.join(path_best_folder, str(model))):
        os.makedirs(os.path.join(path_best_folder, str(model)))
    model_path=os.path.join(path_best_folder, str(model))
    percorso_file = os.path.join(model_path, 'parameters.json')
    df_model=df[df['model']== model]
    best_idx=df_model['best_f1'].idxmax()
    best_parameter=df_model.loc[best_idx]

    parameters = {
        'model': str(best_parameter.model),
        'iteration': int(best_parameter.iteraction),
        'learning_rate': float(best_parameter.learning_rate),
        'mel_bands': int(best_parameter.mel_bands),
        'optimizer': str(best_parameter.optimizer),
        'hop_length': int(best_parameter.hop_length),
        'best_f1': float(best_parameter.best_f1),
        'best_e': int(best_parameter.best_e)}

    print(parameters)
    with open(percorso_file, 'w') as file:
        json.dump(parameters, file)

```
